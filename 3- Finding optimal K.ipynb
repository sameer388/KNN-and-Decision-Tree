```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
import numpy as np
```

```python
cancer = load_breast_cancer()
x_train,x_test,y_train,y_test = train_test_split(cancer.data,cancer.target,random_state = 0,test_size = 0.2)
```

```python
value_of_k = []
score = []
## take the only odd value so that it not stuck
for i in range(1,26,2):
    clf = KNeighborsClassifier(n_neighbors = i)
    value_of_k.append(i)
    score.append(cross_val_score(clf,x_train,y_train).mean())
```

```python
plt.plot(value_of_k,score,label = 'CrossValScore')
plt.grid()
plt.xlabel('K',fontsize = 20,color = 'b')
plt.ylabel('score',fontsize = 20,color = 'b')
plt.legend()
plt.show()
```

```python
## Let's do the experiment with this graph 

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
plt.plot(value_of_k,score,label = 'CrossValScore')

plt.xlabel('K',fontsize = 20,color = 'g')
plt.ylabel('score',fontsize = 20,color = 'g')
plt.axis([0,26,0.9,0.95])
xt = np.arange(1,26,2)
yt = np.arange(0.90,0.96,0.005)
plt.xticks(xt)
plt.yticks(yt)
plt.grid()
plt.legend()
plt.show()
```
